```python
# 1. gapminder.tsv파일을 읽고 다음 작업을 진행하시오
gapminder = pd.read_csv("pandas_data/gapminder.tsv", sep='\t')
gapminder
```

```python
# -자료 크기, 차원, 컬럼, 타입 등 읽어들인 자료의 구조를 확인하시오
gapminder.shape  #(1704, 6)
type(gapminder)  #pandas.core.frame.DataFrame
gapminder.columns  #Index(['country', 'continent', 'year', 'lifeExp', 'pop', 'gdpPercap'], dtype='object')
gapminder.index  #RangeIndex(start=0, stop=1704, step=1)
gapminder.dtypes  #각 열의 자료형태(object(문자열), int64, float64 등)
gapminder.info()  #파이썬의 str 함수와 비슷한 함수(데이터건수, 열 개수, 구조, 결측값 존재 유무까지)
```

```python
# -국가명, 대륙명, 연도 열을 추출하시오
gapminder[['country','continent','year']]
```

```python
# -lifeExp(기대수명),pop(인구),gdp(소득)의 최대/최소/평균/표준편차
def datainf(x):
    print("최대:", np.max(x))
    print("최소:", np.min(x))
    print("평균:", np.mean(x).round(3))
    print("표준편차:", np.std(x).round(3))
```

```python
datainf(gapminder['lifeExp'].values)
# 최대: 82.603
# 최소: 23.599
# 평균: 59.474
# 표준편차: 12.913

datainf(gapminder['pop'].values)
# 최대: 1318683096
# 최소: 60011
# 평균: 29601212.325
# 표준편차: 106126742.555

datainf(gapminder['gdpPercap'].values)
# 최대: 113523.1329
# 최소: 241.16587650000002
# 평균: 7215.327
# 표준편차: 9854.562
```

```python
# 2. 다음 조건을 만족하는 임의의 데이터프레임을 하나 만든다.
# (1) 열의 갯수와 행의 갯수가 각각 5개 이상이어야 한다.
# (2) 열에는 정수, 문자열, 실수 자료형 데이터가 각각 1개 이상씩 포함

data = {
    "총 등록차": [21803351, 22528295, 23202555, 23677366, 23743391],
    "친환경차": [10942, 25278, 56649, 95001, 116379],
    "전기차": [10855, 25108, 55756, 89918, 106099],
    "수소차": [87, 170, 893, 5083, 10280],
    "급속충전기 보급 수량": [881, 1262, 1760, 2033, 1307]
}
idx = [2016, 2017, 2018, 2019, 2020]
```

```python
elecCar = pd.DataFrame(data, index=idx)
elecCar
```

```python
elecCar['충전기당 전기차 수'] = (elecCar['전기차'] / elecCar['급속충전기 보급 수량']).round(2)
elecCar
```

```python
elecCar['충전기 추가 필요유무']=np.where(elecCar['충전기당 전기차 수']<30, "충분", 
                                np.where(elecCar['충전기당 전기차 수']<60, "부족", "많이부족"))
elecCar
```

```python
elecCar.columns.name = "년도"
elecCar
```

```python
# 3. 다음 데이터프레임에서 지정하는 데이터를 뽑아내거나 처리하라.

data = {
    "국어": [80, 90, 70, 30],
    "영어": [90, 70, 60, 40],
    "수학": [90, 60, 80, 70],
}
columns = ["국어", "영어", "수학"]
index = ["춘향", "몽룡", "향단", "방자"]
df = pd.DataFrame(data, index=index, columns=columns)
df
```

```python
# (1) 모든 학생의 수학 점수를 시리즈로 나타낸다.
df['수학']
type(df['수학'])  #pandas.core.series.Series
```

```python
# (2) 모든 학생의 국어와 영어 점수를 데이터 프레임으로 나타낸다.
df[['국어','영어']]
type(df[['국어','영어']])  #pandas.core.frame.DataFrame
```

```python
# (3) 모든 학생의 각 과목 평균 점수를 새로운 열로 추가한다.
df['평균'] = (np.mean(df, axis=1)).round(2)
df
```

```python
# (4) 방자의 영어 점수를 80점으로 수정하고 평균 점수도 다시 계산한다.
df['영어']['방자'] = 80
df['평균'] = (np.mean(df, axis=1)).round(2)
df
```

```python
# (5) 춘향의 점수를 데이터프레임으로 나타낸다.
df['춘향':'춘향']
# 혹은
df[:1]
```

```python
# (6) 향단의 점수를 시리즈로 나타낸다.
df.T['향단']
```
