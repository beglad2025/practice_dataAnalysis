```python
# 주사위를 5번 던져서 나온 수 집합 = [1,1,3,3,5]
# 1 2 3 4 5 6 이라는 주사위 눈에 대해서 
# 2 0 2 0 1 0 를 결과로 얻고 싶다
np.bincount([1,1,3,3,5], minlength=7)
# bincount는 0부터 센다
# minlength가 7이므로 0~6까지가 카운트 범위가 된다
```

```python
d=np.bincount([1,1,3,3,5], minlength=7)
d[1:]  #0처리
# 이렇게 처리해주면 우리가 원하던 결과인 것이다.
```
