# 열 단위로 데이터 추출하기

# loc 속성으로 행 단위 데이터 추출하기

```python
# loc, iloc 연습
gapminder.loc[1000]
gapminder.loc[[1000]]
```

```python
gapminder.iloc[1000]
gapminder.iloc[[1000]]
```

```python
gapminder.loc[1690:,['country','gdpPercap']]
gapminder.iloc[1690:,[0,-1]]
```

# tail과 loc는 조금 달라요!

```python
# 데이터프레임, 시리즈
gapminder.tail(1)
```

```python
gapminder.loc[1703]
```

# iloc 속성으로 행 단위 데이터 추출하기

```python
# 연속, 떨어져 있는 데이터 추출, 음수 기호
gapminder.iloc[:5,-2:]
```

```python
gapminder.iloc[-5:]
```

# 파이썬 슬라이싱 구문을 조합하여 원하는 데이터 추출하기

```python
# 열, 행 추출
gapminder[2:9:2]
```

```python
gapminder[gapminder.year==2002][2:50:2]
```

# iloc 속성과 range 메서드로 원하는 데이터 추출하기

```python
gapminder.iloc[range(0, len(gapminder), 2)]
```

# 열 지정값에 파이썬 슬라이싱을 사용하여 원하는 데이터 추출하기

```python
gapminder['pop'][2:50:2]
```

# loc, iloc 자유자재로 사용하기

```python
gapminder.loc[gapminder.lifeExp>80, ["year","gdpPercap"]]
```

```python
gapminder.iloc[1690:, range(0,6,2)]
```

# 그룹화한 데이터의 개수 세어보기

```python
# 국가별
gapminder.groupby("country").size()
```

```python
# 대륙별
gapminder.groupby("continent").size()
```

```python
# 연도별
gapminder.groupby("year").size()
```
