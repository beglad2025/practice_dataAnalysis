```python
# 1. 파이썬 문제
# - 입력받은 정수가 음수인지 양수인지 0인지 알려주세요.

x = int(input("원하는 수를 입력해주세요: "))
if x>0:
    print(x,"는 양수입니다")
elif x==0:
    print("0입니다")
else:
    print(x, "는 음수입니다")
```

```python
# - 'hello'를 지정한 횟수(0 <= n <= 3)만큼 출력해 보세요.

n = int(input("원하는 반복횟수를 입력하세요"))
if n <= 3 & n >= 0:
    for i in range(n):
        print("hello")
else:
    print("n의 범위는 0이상 3이하 입니다")
```

```python
# 3. iris data 유사 데이터 추출
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris.data
```

```python
iris.data[0,:]  #첫번째 꽃 데이터
```

```python
# 첫번째 꽃 데이터와 가장 유사한 꽃 데이터와 
# 그때의 유클리디안 거리를 출력

# iris.data 표준화
iris_z = (iris.data - np.mean(iris.data, axis=0)) / np.std(iris.data, axis=0)
iris_z
```

```python
distance = np.sqrt(((iris_z[1:len(iris_z)+1] - iris_z[0])**2).sum(axis=1))
distance
```

```python
len(distance)
```

```python
min_distance = distance.min()
min_distance
```

```python
np.where(distance == min_distance)  #16
```
