# plotly, folium : interactive한 시각화 도구

```python
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)  #fig 도화지 한장을 나누지 않고 전체화면에 그래프를 출력
# axis: 축(단수형), axes: 축들(복수형)
```

```python
axes1.hist(tips['total_bill'], bins=10)  #bins=10: 10개의 구간
fig
```

```python
sp=plt.figure()
axes1=sp.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'], tips['tip'])  #산점도 그래프
axes1.set_title("total bill vs tip")
axes1.set_xlabel("total bill")
axes1.set_ylabel("tip")
```

```python
boxp=plt.figure()
axes2=boxp.add_subplot(1,1,1)
axes2.boxplot([tips[tips['sex']=='Female']['tip'],
              tips[tips['sex']=='Male']['tip']])
axes2.set_title("sex")
axes2.set_xlabel("tip")
axes2.set_ylabel("boxplot")
```

```python
# 고차원그래프 표현하는 방법
def cs(a):
    if a=='Female':
        return 0
    else:
        return 1
```

```python
tips['sc'] = tips['sex'].apply(cs)
tips['sc']
```

```python
sp=plt.figure()
axes1=sp.add_subplot(1,1,1)
axes1.scatter(
    x = tips['total_bill'],
    y = tips['tip'],
    r = tips['size'],
    c = tips['sc']
)  #산점도 그래프
axes1.set_title("scatter")
axes1.set_xlabel("total bill")
axes1.set_ylabel("tip")
```
